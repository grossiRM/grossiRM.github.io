```python
import json  ;from datetime import UTC, datetime  ;from pathlib import Path  ;from jinja2 import Environment, FileSystemLoader  
```

```python
with Path("portfolio.json").open(encoding="utf-8") as f:        
    data = json.load(f)
    
data["current_year"]                                            = datetime.now(tz=UTC).year
#for key, value in data.items():                                   print(f"{key}: {value}")
```

```python
data['summary']              # then      data['social_links']
```

```python
if "social_links" in data: 
    for link in data["social_links"]:
        if link.get("svg_path"):
            with Path(link["svg_path"]).open(encoding="utf-8") as svg_file: link["svg_data"] = svg_file.read()

env = Environment(loader=FileSystemLoader("."), autoescape=True)  ;index_template  =env.get_template("index_template.html")        
html_output   = index_template.render(**data)
with Path("index.html").open("w", encoding="utf-8") as f:         f.write(html_output)            
print("HTML file generated successfully!")
```
